```python
import requests
import pandas as pa
from bs4 import BeautifulSoup
r = requests.get("https://en.wikipedia.org/wiki/List_of_causes_of_death_by_rate")
html_contents = r.text
html_soup = BeautifulSoup(html_contents,"lxml")
```

```python
tables = html_soup.find_all('table')
df = pa.read_html(str(tables))[8]
df.head(10)
```
