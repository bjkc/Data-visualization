Now we will be seeing the goal occurance of the player

```python
sns.set(rc={'figure.figsize':(10,10)})
ax = sns.histplot(data = df['Goals'])
ax.set(title = 'Player Goals')
```

Above we can see that not every played have scored the goal as every team member like defenders ,strikers and goal keepers have their own task for their strong team to pass and make the team win.

```python
ax = sns.histplot(data=df, x="Goals", color='green', label="Goals", alpha=.9)
sns.histplot(data=df, x="Assists", color='red', label="Assists", alpha=.8)
ax.set(Title = "Occurence of Goals and Assists",
       xlabel="Goals And Assists")
plt.legend()
plt.show()
```

We can see some kind of correlation above not only scorring every task like defending,assisting the ball who have a chance to score is important too.

  Now we will be looking goals by nationality

```python
ax = df.groupby('nationality').Goals.agg('max').plot(kind = 'bar', title = 'Max World Cup Player Goals Per Nationality')
```

Messiiiiii looks like the argentina had the most goals it has evergreen legend like maradona and lionel messi

Now we will create the box plot of the yellow cards by nationality this card is a warning card and is given before red to give one more chance to players for the fair play.

```python
ax = df.boxplot(column = ('Yel'), by = 'nationality', vert=False)

ax.set_title("")
ax.set_ylabel("Nationality")
ax.set_xlabel("# of Yellow Cards")

plt.show()
```

Above we can see that switzerland and australia have highest yellow cards in their game.

Now we will be creating goals and assists as scatter

```python
ax = sns.scatterplot(data=df, x='Goals', y='Assists', hue='nationality',
                    palette=['pink', 'pink', 'pink', 'pink', 'pink', 'pink', 'pink', 'pink', 'pink', 'pink', 'pink', 'pink', 'pink', 'pink', 'green', 'pink', 'pink', 'pink', 'pink', 'pink', 'pink', 'pink', 'pink', 'pink', 'pink', 'pink', 'pink', 'pink', 'pink'], legend='full')
ax.set_title("Goals and Assists by Nationality")

plt.legend(title = "Nationality")
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

creating stacked barplot of goals and assists of French National Team

```python
df1 = df[['nationality', 'player', 'Goals', 'Assists']]
df1 = (df1[df1['nationality'] == "France"])
ax =df1.plot(kind = 'bar', x='player', stacked = True, alpha=.6)

ax.set(title="French Players Goals and Assists",
             xlabel="Player",
      ylabel="barplot")

plt.show()
```

Now i will be creating corelation heatmap of the whole dataset

```python
ax = sns.heatmap(df.corr(),vmin = -1)
ax.set(title = 'Heatmap of World Cup Player')

plt.show()
```
