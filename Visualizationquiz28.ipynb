```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns 

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')

df.head()
```

```python
ax = sns.scatterplot(data = df, x = 'time', y = 'calories', hue='workout_type', style='workout_type', size="distance")
ax.set(title = "Calories Burned Over Time",
       xlabel = "Time in Seconds",
       ylabel = "Calories Burned",
       xticks = [x for x in range(0,120,20)]) #120 is the max time

plt.legend(title = "Workout Type")
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))

plt.show()
```

```python
sns.set(rc={"figure.figsize":(10,10)})
ax = sns.scatterplot(data = df, 
                x = 'activity_day', 
                y = 'workout_type',
                hue = 'distance',
                hue_order ='time',
                size = 'calories',
                style_order='workout_type',
                x_bins= 10, y_bins= 10,
                legend='auto', n_boot=1000, alpha =1, x_jitter= None,
                ci= None
                )
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
plt.show()
```
