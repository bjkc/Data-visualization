```python
import pandas as pa
import seaborn  as sns

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
df.head()
```

```python
ax = sns.boxplot(x=df['calories'])
```

```python
ax = sns.boxplot(x=df['calories'],y=df['workout_type'])
```

```python
ax = sns.violinplot(x=df['calories'])
```

```python
sns.violinplot(data = df, y = 'workout_type', x = 'calories', cut = 0)
```
